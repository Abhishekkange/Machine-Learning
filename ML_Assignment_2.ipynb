# **Pie Chart**

```python
import matplotlib.pyplot as plt
```

```python
status_count = dataset['furnishingstatus'].value_counts()
slabels = dataset['furnishingstatus'].value_counts().index
plt.pie(status_count,labels=slabels)
```

```python
#add percentage
status_count = dataset['furnishingstatus'].value_counts()
slabels = dataset['furnishingstatus'].value_counts().index
plt.pie(status_count,labels=slabels,autopct='%.2f%%', startangle=90)
plt.title('Furnishing Status Chart')
plt.show()
```

```python
#parking Pie chart
parking_counts = dataset['parking'].value_counts()
plabels = ['one','two','three','four']
plabels
plt.pie(parking_counts,labels=plabels)
plt.title("Parking Distribution")
plt.show()
```

Converting Categorical Data into Numerical Data

```python
dataset['mainroad'].replace(["yes","no"],[1,0],inplace=True)
```

```python
dataset.head()
```

```python
dataset['guestroom'].replace(["yes","no"],[1,0],inplace=True)
dataset['basement'].replace(["yes","no"],[1,0],inplace=True)
dataset['hotwaterheating'].replace(["yes","no"],[1,0],inplace=True)
dataset['airconditioning'].replace(["yes","no"],[1,0],inplace=True)
dataset['prefarea'].replace(["yes","no"],[1,0],inplace=True)
```

```python
dataset.head()
```

```python
dataset['furnishingstatus'].replace(['furnished','semi-furnished','unfurnished'],[2,1,0],inplace=True)
```

```python
dataset.head()
```

Correlation Matrix

```python
matrix_data = dataset.corr()
```

```python
sns.heatmap(matrix_data)
```

annot=True: This parameter, when set to True, adds numeric annotations to each cell in the heatmap, displaying the actual values from the correlation matrix


```python
sns.heatmap(matrix_data,annot=True)
```

```python
#cmap stands for colormap i.e color of Map
sns.heatmap(matrix_data,annot=True,linewidth=0.5,cmap="crest")
```
