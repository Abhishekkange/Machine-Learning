# Pre-Processing

```python
data.insert(9,"Admitted",data['Ad']>0.85)
```

```python
# convert TRUE FALSE to 0 and 1
dataset = data.replace([True,False],[0,1])
```

```python
#drop serial No and AD column
dataset.drop(columns=['Serial No.','Ad'],inplace=True)
```

```python
x = dataset.drop(['Admitted'],axis=1)
```

```python
y = dataset[['Admitted']]
```
