```python
import os
import tarfile
import urllib
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"
def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    os.makedirs(housing_path, exist_ok=True)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()

fetch_housing_data()
```

### Creating a Test set


```python
import numpy as np
def split_train_test(data,test_ratio):
    shuffled_data = np.random.permutation(len(data))
    test_data_size = int(test_ratio* len(data))
    test_indices = shuffled_data[:test_data_size]
    train_indices = shuffled_data[test_data_size:]
    return data.iloc[train_indices],data.iloc[test_indices]


train_set , test_set = split_train_test(df,0.2)


    
```

## Stratified Sampling

```python
df['housing_category'] = pd.cut(df['median_income'],bins=[0.,1.5,3.0,4.5,6.,np.inf],labels=[1,2,3,4,5])
```

```python
df['housing_category'].head()
```

```python
#stratified suffle split 
from sklearn.model_selection import StratifiedShuffleSplit

#creating a split object 
split = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index, test_index in split.split(df,df['housing_category']):      #it split funcation returns index and not value
    strat_train_set = df.iloc[train_index]
    strat_test_set = df.iloc[test_index]

print("Training Set :",strat_train_set.head())
```
