# Model Training 

```python
import pandas as pd
import numpy as np
```

```python
#importing the dataset
housing_data = pd.read_csv('../datasets/housing/housing.csv')
```

```python
# transformation pipelines

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split

#creating a split object 
predicate = housing_data.drop("median_house_value",axis=1)
label  = housing_data['median_house_value']
x_train, x_test,y_train,y_test = train_test_split(predicate,label,test_size=0.2)



#creating a pipeline for numerical data

housing_num = x_train.drop("ocean_proximity",axis=1)

num_pipeline = Pipeline([
    ('fill_null',SimpleImputer(strategy='median')),
    ('std_scalar',StandardScaler()),
])


num_attributes = list(housing_num) # These are column names of all the numerical dat columns
cat_attributes  = ['ocean_proximity']


full_pipeline = ColumnTransformer([

    ('num',num_pipeline,num_attributes),
    ('cat',OneHotEncoder(),cat_attributes),
])

#using full pipeline
housing_prepared = full_pipeline.fit_transform(x_train)
housing_labels = y_train





```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

linear_reg = LinearRegression()
linear_reg.fit(housing_prepared,housing_labels)
housing_data.shape
```

```python
#testing the linear regression model

some_data = housing_data[:2000]
some_data_predicate = some_data.drop("median_house_value",axis=1)
some_data_label = some_data["median_house_value"]

some_data_prepared = full_pipeline.transform(some_data_predicate)


#prediction for one row
output = linear_reg.predict(some_data_prepared)
```
