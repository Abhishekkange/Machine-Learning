```python
import numpy as np
```

### Creating Arrays from List

```python
array_1d = np.array([1,2,3])
print("1D Array :",array_1d)

array_2d = np.array([[1,2,3],[2,4,5]])
print("2D Array :",array_2d)
```

### Zeros Array

```python
zeros_array = np.zeros((3,4))
print(zeros_array)
```

```python
array_ones = np.ones((8,3))
print(array_ones)

#special constant array
print("\n")
special = np.full((2,5),4)
print("\n special array :",special)

#sequence Array
sequence = np.arange(0,12,2)
print("\n sequence Array :",sequence)

#Random array
random_ar = np.random.randint((2,30,42,))
print("\n Random array :",random_ar)
```

## Vector Matrix and tensors


```python
vector = np.array([1,2,3])
print("\n Vector :",vector)

Matrix = np.array([[1,2,3],
                   [3,5,3],
                   [1,5,4]
                   ])
print("\n Matrix :",Matrix)

Tensor = np.array([[[1,2],[3,4]],
                    [[3,4],[6,7]],
                    [[2,2],[4,5]],
                    ])
print("\n Tensor : ",Tensor)
```
