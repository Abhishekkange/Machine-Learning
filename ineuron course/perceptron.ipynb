 ### Perceptron Algorithm

```python
import numpy as np
import pandas as pd
```

```python
X = np.array([[1,2],[2,3],[2,3]])
weights = np.random.randn(X.shape[1])
weights
```

```python
#Perceptron class

class Perceptron:

    def __init__(self,epohcs,size):

        #creates an array of random weights of size of number of input columns
        self.weights = np.random.randn(size)
        self.epohcs = epohcs
        self.learning_rate = 0.1
        self.bias = 0.0


    def activation_function(self,z):
        #compares each value of numpy array and return true or false for given conditio
        result = z>=0
        #returns 1 or 0 for true or false
        return result.astype(int)
        
    
    def _summation_function(self,input_vector,weights):
        return np.dot(input_vector,weights) + self.bias

    def fit(self,X,y):

        #creating input vector out of pandas array 
        input_vector = X.to_numpy()
        
        for ephoc in range(self.epohcs):
            print(f"------- ephoc :{ephoc}/{self.epohcs} -------")
            # all this happens in an ephoc
            result = self._summation_function(input_vector,self.weights)
            y_hat = self.activation_function(result)
            print(f"output after forward pass {y_hat}")
            error = y-y_hat
            print(f"Error  :{error}")
            #learning formula (backpropagation)
            self.weights = self.weights + (self.learning_rate*np.dot(input_vector.T,error))
            
            #bias update rule 
            self.bias = self.bias + self.learning_rate * np.sum(error)
            #weight in current ephoc
            print(f"Weights : {self.weights}")

    def predict(self,X):

        result = self._summation_function(X.to_numpy(),self.weights)
        return self.activation_function(result)
        




```

```python
dataset = pd.read_csv("perceptron_test_data.csv")
```

```python
X = dataset.drop('y',axis=1)
```

```python
y = dataset['y']
```

```python
model = Perceptron(epohcs=100,size=2)
```

```python
model.fit(X,y)
```
