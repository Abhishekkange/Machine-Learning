#### Linear models doesn't work on Non-linear data , so we are visualizing the data of a curve y=X**2 , training a Linear model using that data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor

# Data
X = np.array([[0], [1], [2], [3]])
y = np.array([0, 1, 4, 9])

# ------------------------------
# 1. Linear Regression Model
# ------------------------------
linear_model = LinearRegression()
linear_model.fit(X, y)
linear_pred = linear_model.predict(X)

# ------------------------------
# 2. Neural Network Model
# ------------------------------
nn_model = MLPRegressor(hidden_layer_sizes=(10,), 
                        activation='relu', 
                        solver='adam',
                        max_iter=5000,
                        random_state=42)

nn_model.fit(X, y)
nn_pred = nn_model.predict(X)

# ------------------------------
# Plotting
# ------------------------------
plt.scatter(X, y, label="Original Data", s=80)

plt.plot(X, linear_pred, label="Linear Model", linewidth=3)
plt.plot(X, nn_pred, label="Neural Network", linewidth=3)

plt.title("Linear vs Neural Network on Quadratic Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)

plt.show()
```
