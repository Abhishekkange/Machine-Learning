### Activation Functions 

### Sigmoid Function 

```python
import math

def sigmoid(x):
    return 1/(1+math.exp(x))
```

```python
sigmoid(1)
```

```python
sigmoid(-20)
```

```python
sigmoid(10)
```

### Tanh Function 

```python
def tanh(x):
    return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(-56)
```

### ReLU Function

```python
def relu(x):
    return max(0,x)

```

```python
relu(-15)
```

```python
relu(10)
```
