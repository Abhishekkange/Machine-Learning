# K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
x=dataset1[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
x.head()
```

```python
KmeansModel = KMeans(n_clusters=3)
```

```python
KmeansModel.fit(x)
```

```python
labels = KmeansModel.predict(x)
```
