```python
#filtering the data on certain condition
#matches played in Mumbai
mask = dataset['city']=='Mumbai'
dataset[mask]
```

```python
#making th above part function based
def filterByCity(city):
  mask = dataset['city']==city
  return dataset[mask]
```

```python
filterByCity('Hyderabad').head()
```

```python
#adding multiple mask filter conditions
mask1 = dataset['city']== 'Mumbai'
mask2 = dataset['season']==2017
dataset[mask1 & mask2].head()
```
