Converting Categorical Data into Numerical Data

```python
dataset['mainroad'].replace(["yes","no"],[1,0],inplace=True)
```

```python
dataset.head()
```

```python
dataset['guestroom'].replace(["yes","no"],[1,0],inplace=True)
dataset['basement'].replace(["yes","no"],[1,0],inplace=True)
dataset['hotwaterheating'].replace(["yes","no"],[1,0],inplace=True)
dataset['airconditioning'].replace(["yes","no"],[1,0],inplace=True)
dataset['prefarea'].replace(["yes","no"],[1,0],inplace=True)
```

```python
dataset.head()
```

```python
dataset['furnishingstatus'].replace(['furnished','semi-furnished','unfurnished'],[2,1,0],inplace=True)
```

```python
dataset.head()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
#creating X and Y parameters
x = dataset[['area']]
y = dataset[['price']]
```

```python
x.head()
x.shape
```

```python
train_x,test_x,train_y,test_y = train_test_split(x,y)
```

```python
train_x.head()
```

```python
train_y.head()
```

```python
train_x.shape
```

```python
train_y.shape
```

# Univariate Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
train_x
```

```python
model.fit(train_x,train_y)
```

```python
# import numpy as np
# area_to_predict = np.array([[4000]])
predicted = model.predict(test_x)
```

# **MSR - Mean Square Error **

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(test_y,predicted);
```

```python
mse
```

# **Root Square Mean Error**

```python
rsme = np.sqrt(mse)
```

```python
rsme
```

# Mean Absolute Error

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(test_y,predicted)
```

```python
mae
```

# **R2 Score**

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(test_y,predicted)
```

```python
score
```

# Multi_Variate Linear Regression Model

```python
#creating Attribute and Target sets
x = dataset.drop('price',axis = 1)
```

```python
x.head()
```

```python
y = dataset[['price']]
y.head()
```

Using Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#creating a Simple testing Array
test_value = x_train.iloc[:1]
```

```python
model.predict(test_value)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
predicted_y = model.predict(x_test)
```

# **Mean Square Error**

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test,predicted_y)
```

```python
mse
```

# **Root Mean Square Error**

```python
rmse
```

# **Mean Absolute Error**
